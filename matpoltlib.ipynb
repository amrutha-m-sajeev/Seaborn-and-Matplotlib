```python
x=[150,155,159,156]
y=[55,60,57,79]
plt.plot(x,y,linestyle='dashed')
plt.xlabel("height in cm")
plt.ylabel("weight in cm")
plt.title("height-weight graph")
```

```python
x=[150,155,159,156]
y=[55,60,57,79]
plt.plot(x,y)
font1={'family':'serif','color':'red','size':20}
font2={'family':'serif','color':'blue','size':30}
plt.xlabel("height in cm",fontdict=font1)
plt.ylabel("weight in kg",fontdict=font1)
plt.title("height-weight graph",fontdict=font2,loc='center')
```

```python
x=[12,14,16,18,20]
y=[100,120,130,140,150]
plt.plot(x,y)
font1={'family':'serif','color':'green','size':20}
font2={'family':'serif','color':'red','size':30}
plt.xlabel("Velocity in m\s",fontdict=font1)
plt.ylabel("Displacement in m",fontdict=font1)
plt.title("Velocity-Displacement garph",fontdict=font2)
```

```python
#Grid
x=[150,155,159,156]
y=[55,60,57,79]
plt.plot(x,y)
font1={'family':'serif','color':'red','size':20}
font2={'family':'serif','color':'blue','size':30}
plt.xlabel("height in cm",fontdict=font1)
plt.ylabel("weight in kg",fontdict=font1)
plt.title("height-weight graph",fontdict=font2,loc='center')
plt.grid(color='green',linestyle='--',linewidth=2)
```

```python
#subplot
#it devived the graphpaper into 2 parts
x1=[1,2,3,4]
y1=[8,9,10,11]
plt.subplot(1,2,1)
plt.plot(x1,y1)

x2=[2,4,6,7]
y2=[9,8,7,6]
plt.subplot(1,2,2)#frist numbers decide the number of columns and row means graph generating model,amd 3rd number descide the postition
plt.plot(x2,y2)
```
